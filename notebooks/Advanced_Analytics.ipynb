```python
# Sort data
nav_history = nav_history.sort_values(["amfi_code", "date"])

# Daily return
nav_history["daily_return"] = (
    nav_history.groupby("amfi_code")["nav"]
    .pct_change()
)

nav_history.head()
```

```python
returns = nav_history.dropna(subset=["daily_return"])
```

```python
results = []

for fund, group in returns.groupby("amfi_code"):

    r = group["daily_return"]

    # Historical VaR (95%)
    var95 = r.quantile(0.05)

    # Conditional VaR
    cvar95 = r[r <= var95].mean()

    results.append({
        "amfi_code": fund,
        "VaR_95": var95,
        "CVaR_95": cvar95
    })

var_cvar = pd.DataFrame(results)
```

```python
var_cvar = var_cvar.merge(
    fund_master[["amfi_code", "scheme_name"]],
    on="amfi_code",
    how="left"
)
```

```python
var_cvar = var_cvar[
    [
        "amfi_code",
        "scheme_name",
        "VaR_95",
        "CVaR_95"
    ]
]
```

```python
nav_history = nav_history.sort_values(["amfi_code", "date"])

nav_history["daily_return"] = (
    nav_history.groupby("amfi_code")["nav"]
    .pct_change()
)
```

```python
import numpy as np

rolling_data = []

for fund, group in nav_history.groupby("amfi_code"):

    group = group.sort_values("date").copy()

    rolling_mean = group["daily_return"].rolling(window=90).mean()

    rolling_std = group["daily_return"].rolling(window=90).std()

    group["rolling_sharpe"] = (
        rolling_mean / rolling_std
    ) * np.sqrt(252)

    rolling_data.append(group)

rolling_sharpe = pd.concat(rolling_data)
```

```python
rolling_sharpe = rolling_sharpe.merge(
    fund_master[["amfi_code", "scheme_name"]],
    on="amfi_code",
    how="left"
)
```

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

first_txn = (
    transactions
    .groupby("investor_id")["transaction_date"]
    .min()
    .reset_index()
)

first_txn["cohort_year"] = (
    first_txn["transaction_date"]
    .dt.year
)
```

```python
transactions = transactions.merge(
    first_txn[
        ["investor_id", "cohort_year"]
    ],
    on="investor_id",
    how="left"
)
```

```python
avg_investment = (
    transactions
    .groupby("cohort_year")["amount_inr"]
    .mean()
    .reset_index()
)

avg_investment.columns = [
    "cohort_year",
    "avg_investment"
]

avg_investment
```

```python
total_investment = (
    transactions
    .groupby("cohort_year")["amount_inr"]
    .sum()
    .reset_index()
)

total_investment.columns = [
    "cohort_year",
    "total_investment"
]

total_investment
```

```python
transactions = transactions.merge(
    fund_master[
        ["amfi_code", "scheme_name"]
    ],
    on="amfi_code",
    how="left"
)
```

```python
top_fund = (
    transactions
    .groupby(
        ["cohort_year", "scheme_name"]
    )
    .size()
    .reset_index(name="txn_count")
)

top_fund = (
    top_fund
    .sort_values(
        ["cohort_year", "txn_count"],
        ascending=[True, False]
    )
    .groupby("cohort_year")
    .head(1)
)

top_fund
```

```python
cohort_analysis = (
    avg_investment
    .merge(
        total_investment,
        on="cohort_year"
    )
    .merge(
        top_fund[
            [
                "cohort_year",
                "scheme_name"
            ]
        ],
        on="cohort_year"
    )
)

cohort_analysis
```

```python
cohort_analysis.to_csv(
    "cohort_analysis.csv",
    index=False
)
```

```python
sip_data = transactions[
    transactions["transaction_type"] == "SIP"
].copy()

sip_data.head()
```

```python
sip_data = sip_data.sort_values(
    ["investor_id", "transaction_date"]
)
```

```python
sip_data["gap_days"] = (
    sip_data
    .groupby("investor_id")["transaction_date"]
    .diff()
    .dt.days
)

sip_data.head()
```

```python
sip_count = (
    sip_data
    .groupby("investor_id")
    .size()
    .reset_index(name="sip_count")
)

eligible = sip_count[
    sip_count["sip_count"] >= 6
]

eligible.head()
```

```python
avg_gap = (
    sip_data
    .groupby("investor_id")["gap_days"]
    .mean()
    .reset_index()
)

avg_gap.head()
```

```python
sip_analysis = eligible.merge(
    avg_gap,
    on="investor_id"
)

sip_analysis.head()
```

```python
sip_analysis["risk_status"] = np.where(
    sip_analysis["gap_days"] > 35,
    "At-Risk",
    "Healthy"
)

sip_analysis.head()
```

```python
sip_analysis = sip_analysis.rename(
    columns={
        "gap_days": "average_gap_days"
    }
)

sip_analysis.head()
```

```python
sip_analysis.to_csv(
    "sip_continuity_analysis.csv",
    index=False
)
```

```python
# Convert percentage to decimal
holdings["weight_decimal"] = holdings["weight_pct"] / 100

# Square the weights
holdings["weight_square"] = holdings["weight_decimal"] ** 2

# Calculate HHI
hhi = (
    holdings
    .groupby("amfi_code")["weight_square"]
    .sum()
    .reset_index()
)

hhi.rename(
    columns={"weight_square": "HHI"},
    inplace=True
)

hhi.head()
```

```python
hhi = hhi.merge(
    fund_master[
        ["amfi_code", "scheme_name"]
    ],
    on="amfi_code",
    how="left"
)

hhi = hhi[
    [
        "amfi_code",
        "scheme_name",
        "HHI"
    ]
]

hhi.head()
```

```python
hhi = hhi.sort_values(
    by="HHI",
    ascending=False
)

hhi.head(10)
```

```python
hhi.to_csv(
    "sector_hhi_report.csv",
    index=False
)
```

# Key Insights

### 1. Historical Risk
Funds with higher VaR and CVaR exhibit greater downside risk, indicating larger expected losses during adverse market conditions.

### 2. Rolling Sharpe Ratio
Large-cap funds such as Axis Bluechip, SBI Bluechip, and HDFC Top 100 generally maintained more stable rolling Sharpe ratios over time, indicating relatively consistent risk-adjusted performance.

### 3. Investor Cohorts
Recent investor cohorts contributed a substantial share of investments, while the most preferred schemes varied across cohorts based on transaction frequency.

### 4. SIP Continuity
The SIP continuity analysis identified a significant proportion of investors with an average SIP gap exceeding 35 days, classifying them as At-Risk. This highlights opportunities for investor engagement and retention.

### 5. Portfolio Concentration
Funds with higher HHI values have more concentrated portfolios, implying greater exposure to a limited number of holdings or sectors, whereas lower HHI values indicate better diversification.
